```python
import pandas as pd
```

```python
df = pd.read_excel("SÍNTESE GERAL DA REDE ESTADUAL DE ENSINO.xlsx", header=1)
df.columns = ["indicador", "quant"]
# filtrar só linhas sem subníveis (ex: sem "1.1", "1.1.1"...)
resumo = df[df["indicador"].str.match(r"^\d+\.", na=False)]
```

```python
df["indicador"] = df["indicador"].str.replace("Quantidade de ", "", regex=False)
```

```python
df.columns = ["indicador", "quant"]
df = df.dropna(subset=["quant"])
df = df[df["quant"].astype(str).str.strip() != ""]

# Remove frases repetitivas
df["indicador"] = (
    df["indicador"]
    .str.replace("Quantidade de ", "", regex=False)
    .str.replace("quantidade de ", "", regex=False)
    .str.strip()
)

# Converte para string compacta sem índice
contexto = df.to_csv(index=False, sep="|")
```
